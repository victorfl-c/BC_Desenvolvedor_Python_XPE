#### Funções Aritméticas: Multiplicação

```python
# Multiplicação
print("Multiplicação com dois arrays:\n", x * y) # elemento a elemento
print("Multiplicação com float/int:\n", x * 2) # broadcasting
```

```python
# Multiplicação Matricial
print("Multiplicação matricial (np.dot):\n", np.dot(x, y))
print("Multiplicação matricial (@):\n", x @ y)
print("Multiplicação matricial (.dot):\n", x.dot(y))
```

```python
'''
Exemplo
Solução de um sistema de equações:
    a + 2*b = 7 ==
  1*a + 2*b = 7
  3*a - 2*b = -11
  solução analítica: (a,b) = (-1, 4)
Matricialmente, este problema tem a seguinte forma:
  Ax = c, onde:
  - x = [a, b]
  - A = [[1, 2], [3, -2]]
  - c = [7, -11]
  solução numérica: x = inv(A) @ c,
'''

# Definição do problema
A = np.array([[1, 2], [3, -2]])
c = np.array([[7], [-11]])
print("A:\n", A)
print("c:\n", c)
print("inv(A):\n", np.linalg.inv(A)) # linalg.inv() retorna a inversa de uma matriz
```

```python
# Solução
# x = np.linalg.inv(A) @ c
x = np.dot(np.linalg.inv(A), c)
print("(a, b):\n", x)
print("(a, b):", x.ravel())
```

### Conceitos básicos de Regressão linear

Um problema de regressão linear é determinar uma função ou uma reta que melhor descreve o comportamento de um conjunto de dados.


*   A escolha da melhor função deve ser baseada em um critério
*   Um critério comumente utilizada é o erro quadrático, que queremos minimizar;
*   Com o critério definido e, sabendo que a função linear possui o seguinte formato: y = f(x) = ax + b
*   O problema de regressão resume-se à determinação dos coeficientes a e b, visto que x e y são dados de entrada.








```python
# visualização de dados
import matplotlib.pyplot as plt
```

```python
# dados
x = np.array([-1, -0.777777778, -0.555555556, -0.33333333,-0.11111111,
     0.11111111, 0.333333333, 0.55555556, 0.77777778, 1.])

y = np.array([-1.13695485, -0.57156547, -0.21569874, 0.56987651, 0.45698756,
     1.14569856, 1.64598796, 2.17895436, 2.64589756, 2.95698741])
```

```python
# plot dos dados
plt.figure(figsize=(10, 5))
plt.plot(x, y, 'o', label='dados originais')
plt.legend() # expor dados originais no plot
plt.xlabel("x")
plt.ylabel("y")
plt.grid() # define o grid do plot
plt.show()
```

```python
"""
Iremos estimar uma função do tipo: y = a*x + b
ou seja, devemos achar quais valores de a e b
que melhore representam os dados


Os valores reais de a e b são: (a, b): 2, 1
"""
# transformando pra numpy e vetor coluna
x, y = np.array(x).reshape(-1, 1), np.array(y).reshape(-1, 1)
# o -1 significa que quero ignorar a quantidade de linhas, e o 1 forçar a formar
# um array de apenas 1 coluna

# adicionar bias: para estimar o termo b
X = np.hstack((x, np.ones(x.shape)))

# estimando a e b
beta = np.linalg.pinv(X).dot(y)
print("a estimado:", beta[0][0])
print("b estimado:", beta[1][0])
```

```python
# plot dos dados
plt.figure(figsize=(10, 5))
plt.plot(x, y, 'o', label='dados originais')
plt.plot(x, X.dot(beta), label='regressão linear')
plt.legend() # expor dados originais no plot
plt.xlabel("x")
plt.ylabel("y")
plt.title("Regressão linear com numpy")
plt.grid() # define o grid do plot
plt.show()
```

# Scikit-Learn

## Como implementar um classificador de padrões no Scikit-learn

```python
df.head()
```

```python
# extração de x e y
x, y = df[['temperatura']].values, df[['classification']].values
print('x:\n', x)
print('y:\n', y)
```

```python
# pré-processamento
from sklearn.preprocessing import LabelEncoder
```

```python
# conversão de y para valores numéricos
le = LabelEncoder()
y = le.fit_transform(y.ravel())
print('y:\n', y)
```

```python
# modelo
from sklearn.linear_model import LogisticRegression
```

```python
# Classificador
clf = LogisticRegression()
clf.fit(x, y)
```

```python
# gerando 100 valores de temperatura
# linearmente espaçados entre 0 e 45
# predição em novos valores de temperatura
x_test = np.linspace(start=0., stop=45., num=100).reshape(-1, 1)

# predição desses valores
y_pred = clf.predict(x_test)
```

```python
print(y_pred)
```

```python
# conversão de y_pred para os valores originais
y_pred = le.inverse_transform(y_pred)
```

```python
# output
output = {'new_temp': x_test.ravel(),
          'new_class': y_pred.ravel()}
output = pd.DataFrame(output)
```

```python
output.head()
```

```python
# estatisticas
output.info()
```

```python
# estatisticas
output.describe()
```

```python
# contagem de valores gerados
output['new_class'].value_counts().plot.bar(figsize=(10, 5),
                                            rot=0,
                                            title='# de novos valores gerados');
```

```python
# distribuição do output produzido
# conseguimos inferir a classificação novas temperaturas
# a partir de um dataset com 6 exemplos
output.boxplot(by='new_class', figsize=(10, 5));
```

```python
# sistema automático
def classify_temp():
  '''classifica o input do usuário'''

  ask = True
  while ask:
    # input de temperatura
    temp = input("Insira a temperatura (graus Celsius): ")

    # transformar para numpy array
    temp = np.array(float(temp)).reshape(-1, 1)

    # realiza classificação
    class_temp = clf.predict(temp)

    # transformação inversa para retornar a string original
    class_temp = le.inverse_transform(class_temp)

    # classificação
    print(f'A classificação da temperatura {temp.ravel()[0]} é: {class_temp[0]}')

    # perguntar
    ask = input("Nova classificação (y/n): ") == 'y'
```

```python
# rodando programa
classify_temp()
```

## Regressão Linear no Scikit-Learn I

```python
# dados
x = np.array([-1, -0.777777778, -0.555555556, -0.33333333,-0.11111111,
     0.11111111, 0.333333333, 0.55555556, 0.77777778, 1.])

y = np.array([-1.13695485, -0.57156547, -0.21569874, 0.56987651, 0.45698756,
     1.14569856, 1.64598796, 2.17895436, 2.64589756, 2.95698741])
```

```python
# plot dos dados
plt.figure(figsize=(10, 5))
plt.plot(x, y, 'o', label='dados originais')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
# transformando em numpy array
x, y = np.array(x).reshape(-1, 1), np.array(y).reshape(-1, 1)
```

```python
# modelo
from sklearn.linear_model import LinearRegression
```

```python
# treinando o modelo: y = a*x + b, valores reais (a, b) = (2, 1)
reg = LinearRegression()
reg.fit(x, y)
```

```python
# coeficientes a, b estimados:
# valores estimados usando o numpy diretamente
# a estimado no numpy: 2.05414951
# b estimado no numpy: 0.96798926
print('a estimado: ', reg.coef_.ravel()[0]) # termo dependente = a
print('b estimado: ', reg.intercept_.ravel()[0]) # termo independente = b
```

```python
# predição do modelo
y_pred = reg.predict(x)
```

```python
# score do modelo
score = reg.score(x, y)
print('Score: ', score)
```

```python
# plot dos dados
plt.figure(figsize=(10, 5))
plt.plot(x, y, 'o', label='dados originais')
plt.plot(x, y_pred, label=f'regressão linear (R2: {score:.3f})')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title("Regressão Linear no scikit-learn")
plt.grid()
plt.show()
```

```python
# função para cáculo do MSE
def mse(y_true, y_pred, is_ref = False):

  #mse modelo
  if is_ref:
    mse = ((y_true - y_true.mean())**2).mean()
  else:
    mse = ((y_true - y_true)**2).mean()

  return mse
```

```python
# função para cálculo do coeficiente de determinação R2
def r2(mse_reg, mse_ref):
  return 1 - mse_reg/mse_ref
```
